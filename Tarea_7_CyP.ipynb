```
# PROBLEMA 4.11. PÁG.274. PRIMER INTENTO

#Tn, Tc, Ts, son las varibles que almacenaran el total de paises en Norte America, Centro y Sur respectivamente
Tn= 0
Tc= 0
Ts= 0
print("PAÍSES EN AMÉRICA ")
Tn= int (input("Ingrese el total de paises del Norte: "))
Tc= int (input("Ingrese el total de paises del Centro: "))
#Ta= Total de países en América
Ts= int (input("Ingrese el total de paises del Sur: "))
Ta= Tn + Tc + Tc + 3
if 1<Tn<30 and 1<Tc<30 and 1<Ts<30:
  #Declaramos los arreglos unidimensionales
  nom_pais= "Pais" 
  NORTE=[nom_pais for x in range(Tn)]
  CENTRO=[nom_pais for x in range(Tc)]
  SUR= [nom_pais for x in range(Ts)]
  AMERICA= [nom_pais for x in range(Ta)]

  print('''
  Ingrese el nombre de los países de Norteamérica''')
  for i in range (Tn):
    NORTE[i]= str (input(f"Ingresa el nombre del país {i+1}: "))

  print('''
  Ingrese el nombre de los países de Centroamérica''')
  for i in range (Tc):
    CENTRO[i]= str (input(f"Ingresa el nombre del país {i+1}: "))

  print('''
  Ingrese el nombre de los países de Sudamérica''')
  for i in range (Tn):
    SUR[i]= str (input(f"Ingresa el nombre del país {i+1}: "))

  PA= 0
  PN= 0
  PC= 0
  PS= 0
  while PN!=(Tn) and PC!=(Tc) and PS!=(Ts):
    PA= PA + 1
    if (SUR[PS]<CENTRO[PC]) and SUR[PS]<NORTE[PN]:
       AMERICA[PA]= SUR[PS]
       PS= PS + 1
    else:
      if CENTRO[PC]<NORTE[PN]:
         AMERICA[PA]= CENTRO[PC]
         PC= PC + 1
      else:
         AMERICA[PA]= NORTE[PN]
         PN= PN + 1 

  while (PC!=Tc) and (PS!=Ts):
    PA= PA + 1
    if SUR[PS]<CENTRO[PC]:
       AMERICA[PA]= SUR[PS]
       PS= PS + 1
    else: 
       AMERICA[PA]= CENTRO[PC]
       PC= PC + 1

  while (PN != Tn) and (PS != Ts):
    PA= PA + 1
    if SUR[PS] < NORTE[PN]:
       AMERICA[PA]= SUR[PS]
       PS= PS + 1
    else: 
       AMERICA[PA]= NORTE[PN]
       PN= PN + 1 

  while (PN!=Tn) and (PC!=Tc):
    PA= PA + 1
    if CENTRO[PC] < NORTE[PN]:
       AMERICA[PA]= CENTRO[PC]
       PC= PC + 1
    else:
       AMERICA[PA]= NORTE[PN]
       PN= PN + 1 
  while PS!=(Ts):
    PA= PA + 1
    AMERICA [PA]= SUR[PS]
    PS= PS + 1

  while PC!= (Tc):
   PA= PA + 1
   AMERICA[PA]= CENTRO[PC]
   PC= PC + 1

  while PN!= (Tn):
   PA= PA + 1    
   AMERICA[PA]= NORTE[PN]
   PN= PN + 1 

  print('''------------------------------ 
Arreglo America:''')
  for i in range(1, PA+1):
    print(AMERICA[i])



else:
  print("Verifique el número de Países ingresados.")
```
